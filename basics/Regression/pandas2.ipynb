```python
import pandas as pd

df = pd.read_csv(
    "/home/jjk339/learn/learn-pytorch/Basics/Regression/data/used_cars.csv"
)
df.head()
```

```python
# Convert the 'price' column to integers
price = df["price"]
price = price.str.replace("$", "")
price = price.str.replace(",", "")
price = price.astype(int)

df["price"] = price
df.head()
```

```python
df["price"].mean()
```

```python
df_hyundai = df[df["brand"] == "Hyundai"]
print(df_hyundai["price"].mean())

df_ford = df[df["brand"] == "Ford"]
print(df_ford["price"].mean())
```

```python
# This function creates dummy variables for the 'brand' column
pd.get_dummies(df["brand"])
```

```python
# This function will give the correlation between 'model_year' and 'price'
df[["model_year", "price"]].corr()
```
